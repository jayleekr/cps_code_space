자 그럼 이 데이터들을 조금 더 정돈된 data structure로 담아 봅시다.
먼저 우리가 배운 dict로 담아 json으로 export 할 수 있겠죠?

```python
# print(fulldata_df.iloc[3])

# Convert the dateframes to dictionaries with UTF-8 encoding and below format

#   { 
#       sub_no : "102",
#   }
```

```python
# Get the first row of fulldata_df
print(fulldata_df.iloc[0])

# Define a class for each row of the fulldata_df
class SubjectData:
    def __init__(self, sub_no, set_no, orig_sent_num, original_sent, recall_sent_num,
                 recall_sent, age_group_clf, sex_clf, cesd_boston_clf, edu_years, recall_total,
                 mmse, fab, span_task, speed_task, vocab_task, sequencing_task, fluency_p,
                 fluency_c, cesd_boston, recall_prop_count, recall_prop_ratio):
        self.sub_no = sub_no
        self.set_no = set_no
        self.orig_sent_num = orig_sent_num
        self.original_sent = original_sent
        self.recall_sent_num = recall_sent_num
        self.recall_sent = recall_sent
        self.age_group_clf = age_group_clf
        self.sex_clf = sex_clf
        self.cesd_boston_clf = cesd_boston_clf
        self.edu_years = edu_years
        self.recall_total = recall_total
        self.mmse = mmse
        self.fab = fab
        self.span_task = span_task
        self.speed_task = speed_task
        self.vocab_task = vocab_task
        self.sequencing_task = sequencing_task
        self.fluency_p = fluency_p
        self.fluency_c = fluency_c
        self.cesd_boston = cesd_boston
        self.recall_prop_count = recall_prop_count
        self.recall_prop_ratio = recall_prop_ratio
    
    def to_json(self):
        return json.dumps(self, default=lambda o: o.__dict__, ensure_ascii=False, indent=4)
    
    
# Create a list of SubjectData objects
subject_data_list = []
for index, row in fulldata_df.iterrows():
    subject_data = SubjectData(sub_no=row["sub_no"],
                               set_no=row["set_no"],
                               orig_sent_num=row["orig_sent_num"],
                               original_sent=row["original_sent"],
                               recall_sent_num=row["recall_sent_num"],
                               recall_sent=row["recall_sent"],
                               age_group_clf=row["age_group_clf"],
                               sex_clf=row["sex_clf"],
                               cesd_boston_clf=row["cesd_boston_clf"],
                               edu_years=row["edu_years"],
                               recall_total=row["recall_total"],
                               mmse=row["mmse"],
                               fab=row["fab"],
                               span_task=row["span_task"],
                               speed_task=row["speed_task"],
                               vocab_task=row["vocab_task"],
                               sequencing_task=row["sequencing_task"],
                               fluency_p=row["fluency_p"],
                               fluency_c=row["fluency_c"],
                               cesd_boston=row["cesd_boston"],
                               recall_prop_count=row["recall_prop_count"],
                               recall_prop_ratio=row["recall_prop_ratio"])
    subject_data_list.append(subject_data)


for subject_data in subject_data_list:
    print("=========================================")
    print(subject_data.to_json())
```
