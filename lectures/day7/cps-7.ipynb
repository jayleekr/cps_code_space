```python
import json
import numpy as np
# Checkif pandas is installed
try:
    import pandas 
except ImportError:
    # Run pip install pandas
    import pip
    pip.main(['install', 'pandas'])
    import pandas 
try:
    import matplotlib.pyplot as plt
except ImportError:
    # Run pip install matplotlib
    import pip
    pip.main(['install', 'matplotlib'])
    import matplotlib.pyplot as plt


# Read data from 1_demo_update.xlsx
df = pandas.read_excel('1_demo_update.xlsx')
print("==========================================")
print(df)

```

```python
# - Scatter plot
# - Line plot
# - Bar plot
# - Histogram
# - Heatmap
# - Violin plot
# - Area plot
# - Pie chart
# - Bubble plot
# - Treemap

# age vs depression score

# depression score can be calculated as follows:
# depressionScoreTotal = S-CESD * 2 + CESD


# 1. Scatter plot
x = df['age']
y = df['S-CESD'].fillna(0) * 2 + df['CESD'].fillna(0)


print(y)
print(f"average depression score: {np.mean(y)}")
print(f"max depression score: {np.max(y)}")

# # scatter argues: x, y, color, alpha, marker, s, linewidths, edgecolors, label, cmap, c
# # alpha: transparency
# # marker: o, v, ^, <, >, s, p, *, h, H, D, d, P, X
# # s: size
# # linewidths: line width
# # edgecolors: edge color
# # cmap: color map
# # c: color
# # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

plt.scatter(x, y, c=x, cmap='Set1',  alpha=0.5, marker='*', s=50, linewidths=1, 
            edgecolors='black', label='depression score')
plt.title('age vs depression score')
plt.xlabel('age')
plt.ylabel('depression score')
plt.show()
```

```python
# 2. Line plot
# line plot is similar to scatter plot
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html


# Fit a polynomial regression model of degree 1
p = np.polyfit(x, y, 1)
print(f"polynomial regression model  of degree 2: {p}")
plt.plot(x, np.poly1d(p)(x), 'r--', label='polynomial regression model of degree 1')
plt.scatter(x, y, c=x, cmap='Set1',  alpha=0.5, marker='*', s=50, linewidths=1,
            edgecolors='black', label='depression score')
plt.title('age vs depression score')
plt.xlabel('age')
plt.ylabel('depression score')
plt.legend()
plt.show()
```

```python



# Fit a polynomial regression model of degree 2
p = np.polyfit(x, y, 2)
print(f"polynomial regression model of degree 2: {p}")
plt.plot(x, np.poly1d(p)(x), 'r--', label='polynomial regression model of degree 2')
plt.scatter(x, y, c=x, cmap='Set1',  alpha=0.5, marker='*', s=50, linewidths=1,
            edgecolors='black', label='depression score')
plt.title('age vs depression score')
plt.xlabel('age')
plt.ylabel('depression score')
plt.legend()
plt.show()
```

```python

# Fit a polynomial regression model of degree 3
p = np.polyfit(x, y, 3)
print(f"polynomial regression model of degree 3: {p}")
plt.plot(x, np.poly1d(p)(x), 'r--', label='polynomial regression model of degree 3')
plt.scatter(x, y, c=x, cmap='Set1',  alpha=0.5, marker='*', s=50, linewidths=1,
            edgecolors='black', label='depression score')
plt.title('age vs depression score')
plt.xlabel('age')
plt.ylabel('depression score')
plt.legend()
plt.show()
```

```python

# Regression with scipy
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html

from scipy import stats

# Linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
print(f"slope: {slope}")
print(f"intercept: {intercept}")
print(f"r_value: {r_value}")
print(f"p_value: {p_value}")
print(f"std_err: {std_err}")
plt.plot(x, slope*x+intercept, 'r--', label='linear regression')
plt.scatter(x, y, c=x, cmap='Set1',  alpha=0.5, marker='*', s=50, linewidths=1,
            edgecolors='black', label='depression score')
plt.title('age vs depression score')
plt.xlabel('age')
plt.ylabel('depression score')
plt.legend()
plt.show()
```

```python

# Non-linear regression
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * np.exp(-b * x) + c

popt, pcov = curve_fit(func, x, y)
print(f"popt: {popt}")
print(f"pcov: {pcov}")

plt.plot(x, func(x, *popt), 'r--', label='non-linear regression')
plt.scatter(x, y, c=x, cmap='Set1',  alpha=0.5, marker='*', s=50, linewidths=1,
            edgecolors='black', label='depression score')
plt.title('age vs depression score')
plt.xlabel('age')
plt.ylabel('depression score')
plt.legend()
plt.show()



```

```python
# 3. Bar plot
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

# Bar plot of depression score
plt.bar(x, y, color='blue', alpha=0.5, label='depression score')
plt.title('age vs depression score')
plt.xlabel('age')
plt.ylabel('depression score')
plt.legend()
plt.show()
```

```python
# Bar plot of depression score with error bar
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.errorbar.html

# In a plot, an error bar is a line representing the range of values for a data point, typically the range of uncertainty for a mean value or a standard deviation. It can be useful to show the variability of data or the accuracy of measurements.

# Calculate the standard deviation of depression score
std = np.std(y)
print(f"standard deviation of depression score: {std}")

# Plot the bar plot
plt.bar(x, y, color='blue', alpha=0.5, label='depression score')

# Plot the error bar
plt.errorbar(x, y, yerr=std, fmt='o', color='black', ecolor='black', elinewidth=1, capsize=2, label='error bar')
```

```python
# 4. Histogram
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

# Histogram of depression score

counts, bins = np.histogram(y, bins=10)
print(f"counts: {counts}")
print(f"bins: {bins}")

plt.stairs(counts, bins, fill=True, label='depression score')
plt.xlabel('depression score')
plt.ylabel('frequency')
plt.title('Histogram of depression score')
plt.legend()
plt.show()

```

```python
#  5. Box plot of age vs depression score
plt.boxplot(y, labels=['depression score'], showmeans=True)
plt.title('Box plot of depression score')
plt.show()
```

```python
# 6. Violin plot of age vs depression score
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html

# Easy example
plt.violinplot(y, showmeans=True)
plt.title('Violin plot of depression score')
plt.show()


# Very difficult example :D 
# data[(n, n_age)] : depression score of age of n*10 to (n+1)*10, with n*10 age
data=[]
for i in range(1, 10):
    df_of_nth_age=df[(df['age'] > i*10) & (df['age'] <= (i+1)*10)]
    # calculate depression score of nth age
    # depression_total =  S-CESD * 2 + CESD
    if df_of_nth_age.empty:
        continue
    dep_total_of_nth_age = df_of_nth_age['S-CESD'].fillna(0) * 2
    + df_of_nth_age['CESD'].fillna(0)
    data.append((dep_total_of_nth_age, i*10))

plt.violinplot([d[0] for d in data], [d[1] for d in data], showmeans=True, widths=5
               , showextrema=True, showmedians=True)
plt.title('Violin plot of age vs depression score')
plt.xlabel('age')
plt.ylabel('depression score')
plt.show()

```

```python
# 7. Heatmap
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html
# Correlation matrix
df_delta= df
corr = df_delta.corr()

# Plot the heatmap
plt.imshow(corr, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns)
plt.show()

# it seems like there are too many CESD of nth, let's remove them
for i in range(1, 21):
    df_delta = df_delta.drop(['CESD'+str(i)], axis=1)

plt.imshow(df_delta.corr(), cmap='hot', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(df_delta.corr())), df_delta.corr().columns, rotation='vertical')
plt.yticks(range(len(df_delta.corr())), df_delta.corr().columns)
plt.show()

# It's better now !!
```
